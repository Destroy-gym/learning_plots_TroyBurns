```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data (2).csv")
```

```python
spotify_df.head()
```

```python
#group by genre and do avg total streams
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending = False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")

plt.show
```

```python
plt.figure(figsize=(10,6))

plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45,ha="right")

plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)
#gca is get current axis
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45,ha="right")

plt.tight_layout()
plt.show
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending = False)
avg_artist_streams
```

```python
plt.Figure(figsize=(10,6))
colors = ["#756bb1"]*len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#db36a4"

#plt.bar(avg_artist_streams.index,avg_artist_streams.values,color ="#756bb1", width = 0.8)
plt.bar(avg_artist_streams.index,avg_artist_streams.values,color = colors, width = 0.8)
#color brewer is a good place for color hexcode
#can also just put "pink" in place of hexcode
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45,ha="right")

plt.tight_layout()
plt.show
```

```python
plt.Figure(figsize=(10,6))

special_artist = "Ariana Grande"
colors = ["#db36a4" if artist == special_artist else"gray"for artist in avg_artist_streams.index]


#plt.bar(avg_artist_streams.index,avg_artist_streams.values,color ="#756bb1", width = 0.8)
plt.bar(avg_artist_streams.index,avg_artist_streams.values,color = colors, width = 0.9)
#color brewer is a good place for color hexcode
#can also just put "pink" in place of hexcode
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Ariana Grande Ranks Second in Total Overall Streams")
plt.xticks(rotation=45,ha="right")

plt.tight_layout()
plt.show
```

```python
plt.scatter(spotify_df["Streams Last 30 Days (Millions)"],spotify_df["Total Streams (Millions)"])
plt.show
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"],
    spotify_df["Total Streams (Millions)"],
    color = "#2d9b11",
    alpha= 0.8,
    s=25)
#alpha is the transperancy of the dots
#s is the size of the dots
plt.xlabel("Streams This Month")
plt.ylabel("Total Streams")
plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(10,6))

featured_genre = "Pop"

colors = ["#2d9b11" if genre == featured_genre else "#bbd4c3" for genre in spotify_df["Genre"]]


plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"],
    spotify_df["Total Streams (Millions)"],
    color = colors,
    alpha= 0.8,
    s=25)
#alpha is the transperancy of the dots
#s is the size of the dots
plt.xlabel("Streams This Month")
plt.ylabel("Total Streams")



ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do Pop Songs Achieve Higher Monthly Listens or Total Streams?", fontsize = 18,pad = 45)
#instead of manually changing all of the values (as seen above) you can do a for loop 
for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

plt.tight_layout()
plt.show()
```

```python
yearly_avg = spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize=(10,6))
plt.plot(yearly_avg.index,
         yearly_avg.values,
         marker='o',
         linestyle = "--",
         color = "#20a32b",
         linewidth = 0.5)

plt.xlabel("Release Year")
plt.ylabel("Average Total Listens")
plt.title("Did Listeners Stream Less in 2022?", fontsize = 18, pad = 20)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
#add a vertical line on the x axis
ax.axvline(
    2022,
    color = "#05461B",
    linewidth= 0.5,
    alpha = 0.7)

ax.text(
    2022+0.05,
    2225,
    "Dip in Total Streams",
    color = "#05461B",
    fontsize= 8
)

plt.tight_layout()
plt.ylim(0,3500)
plt.show()
```

```python
category = "Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(7,7))
plt.pie(
    platform_totals,
    labels = None,
    startangle=90,
    colors = ["grey","#1D8A23"],
    autopct="%1.1f%%")

plt.title("Near Even Split Between Premium and Free Users")

plt.legend(
    labels = platform_totals.index,
    loc="best",
    title = "Platform"
)
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(8,8))

wedges, texts, autotexts =ax.pie(
    platform_totals,
    autopct="%1.1f%%",
    startangle=110,
    colors = ["grey","#1D8A23"])
 
ax.set_title(
    "Near Even Split Between Premium and Free Users",
    fontsize = 40,
    pad=15,
    loc="center"
)   

for autotext in autotexts:
    autotext.set_fontsize(30)
    autotext.set_color("white")
    autotext.set_weight("bold")

ax.legend(
    wedges,
    platform_totals.index,
    loc="lower center",
    bbox_to_anchor = (0.5,0.95),
    ncol=2
)

plt.tight_layout()
plt.show()
```
